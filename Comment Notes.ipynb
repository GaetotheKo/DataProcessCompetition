```python
2021.05.07. wk.kang

Data Pre-processing 요약
1. Data imbalancing : Fail Sample의 수가 더 많음.
    fail data set을 random sampling하여 pass data set 의 수만큼 만들어서 하나의 batch set으로 사용하고,
    다시 random sampling하여 다른 batch data set으로 사용

2. 불필요한 parameter 제거
    Device Serial Number
    그 외, 데이터는 포함 시켜서 돌려 보면서 parameter importance 결과에 따라 parameter 선정 (XGBoost)

3. Normalize
    각 Parameter의 Min/Max 값을 이용하여 0~1 사이의 값을 갖도록 normalize

4. Erase fail block과 retire block no.는 거의 동일한 의미임으로 erase fail block paramter는 제거 하고, 
    retire block no.만 사용

5. Host Write와 NAND PGM은 WAF 반영 여부가 차이점 임으로, Host Write는 사용하지 않고, NAND PGM만 사용

6. Outlier 제거 (추후 outlier를 간편하게 확인 할 수 있다면 outlier 제거)

7. Validation Set 준비
    Training set의 일부 data를 validation set으로 구분하여 validation 시 사용
    어느 정도를 validation set으로 나눌 것인지는 data set의 수 / pass / fail의 수를 뽑아 보고 결정

8. 하나의 sample에 존재하는 다수의 time series data 활용
    Data set 준비 (다음과 같은 안을 생각해 볼 수 있음)
    1. POH 단위시간(24시간?)당 parameter의 평균 변화율
    맨 첫 값과 맨 마지막 값의 차를 POH로 나눈다
    -> 이 방법은 전체 평균은 알 수 있지만, 시간에 따른 변화량이 달라지는 것은 반영할 수 없다.
    -> 이 방법은 sample당 하나의 data 생성
    2. t-1, t 사이의 각 parameter 변화량을 새로운 데이터로 추가하여 사용
    t-1 data와의 차이를 구하고, t-1, t와의 POH로 나누어서 변화량을 계산
    -> 이 방법은 time step별로 변화량의 변화 반영 가능
    -> 이 방법은 sample당 time step수 -1개의 data 생성
    -> 하나의 sample에 대해 다수의 data가 있을 경우 어떻게 regression 모델을 만들어야 할까?
        각각의 row를 독립적이라고 생각하고 regression하는 방법
        이 경우는 시간 변화에 관한 내용을 반영하기 어려움
        -> LSTM의 경우는 가능 함.
```
